```python
# Create lists of paths for the images and labels
import os
import glob

image_folder = '/content/drive/My Drive/arabidopsis_kmeans/image'
image_paths = sorted(glob.glob(os.path.join(image_folder, '*.jpg')))
print(image_paths)

label_folder = '/content/drive/My Drive/arabidopsis_kmeans/label'
label_paths = sorted(glob.glob(os.path.join(label_folder, '*.png')))
print(label_paths)
```

```python
from datasets import Dataset, DatasetDict, Image
```

```python
# Function to create a custom dataset
def create_dataset(image_paths, label_paths):
    dataset = Dataset.from_dict({"pixel_values": sorted(image_paths),
                                 "label": sorted(label_paths)})
    dataset = dataset.cast_column("pixel_values", Image())
    dataset = dataset.cast_column("label", Image())

    return dataset

# Create Dataset objects
train_dataset = create_dataset(image_paths, label_paths)

# Create DatasetDict
dataset = DatasetDict({
    "train": train_dataset
})
```
